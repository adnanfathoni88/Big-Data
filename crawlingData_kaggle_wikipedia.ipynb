# Beautiful Soup

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
    'AppleWebKit/537.36 (KHTML, like Gecko) '
    'Chrome/117.0 Safari/537.36'
}

page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_='wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
wordTitle = table.find_all('th')
wordTitle
```

```python
wordTableTitle = [title.text.strip() for title in wordTitle]
wordTableTitle

print(wordTableTitle)
```

```python
df = pd.DataFrame(columns=wordTableTitle)
df
```

```python
columnTr = table.find_all('tr')
```

```python
for rows in columnTr[1:]:
  row_data = rows.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]

  length = len(df)
  df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv('companies.csv', index=False)
```
