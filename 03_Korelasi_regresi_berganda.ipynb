# Import Dataset

```python
# import library
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/KULIAH/2024/BIG DATA/dataset/Global_AI_Content_Impact_Dataset.csv');
df.info()
```

Regresi Linear Berganda

```python
# multiple linear regression

np.random.seed(42)
split = np.random.rand(len(df)) < 0.8
df_train = df[split]
df_test = df[~split]

# data train
x_train = np.asanyarray(df_train[['Human-AI Collaboration Rate (%)', 'AI Adoption Rate (%)', 'Revenue Increase Due to AI (%)', 'Consumer Trust in AI (%)' ]])
y_train = np.asanyarray(df_train[['Job Loss Due to AI (%)']])

# data test
x_test = np.asanyarray(df_test[['Human-AI Collaboration Rate (%)', 'AI Adoption Rate (%)', 'Revenue Increase Due to AI (%)', 'Consumer Trust in AI (%)' ]])
y_test = np.asanyarray(df_test[['Job Loss Due to AI (%)']])
```

```python
from sklearn.linear_model import LinearRegression

# membuat model dan melatih model
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)

# menampilkan koefisien dan intercept
print('Koefisien: ', lr_model.coef_)
print('Intercept / konstanta: ', lr_model.intercept_)
```

Job Loss Due to AI (%) = 13.5351 + (0.0660 * Human-AI Collaboration Rate) + (0.0010 * AI Adoption Rate) + (0.0883 * Revenue Increase Due to AI) + (0.0795 * Consumer Trust in AI)

Angka-angka ini (koefisien dan intercept) adalah parameter yang telah "dipelajari" oleh model dari data pelatihan dan akan digunakan untuk membuat prediksi tentang Job Loss Due to AI (%) berdasarkan nilai dari empat variabel independen tersebut.
